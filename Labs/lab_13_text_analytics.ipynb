# Part 3: VADER
### What you are going to do:
- run and train a VADER model
- evaluate its performance

### Why this matters:
Sentiment analysis matters for a business because it helps them understand how customers feel about their products, services, or brand by analyzing large amounts of text data like reviews, social media posts, and comments. By identifying positive or negative sentiment at scale, companies can respond to customer concerns, improve their offerings, and make more informed decisions about marketing, customer experience, and overall strategy.

### How VADER Works
VADER (Valence Aware Dictionary and sEntiment Reasoner) is a rule-based sentiment analysis tool that uses a predefined list (lexicon) of words with associated sentiment scores, along with rules that account for things like capitalization, punctuation, and intensifiers (e.g., “very” or “extremely”). It combines these factors to produce an overall sentiment score indicating whether text is positive, negative, or neutral.

Since it looks at one word at a time it may struggle with longer sentences. So VADER is best suited for analyzing social media and reviews.

**Things to notice:**
- what does the score_to_label function do?

````python
# set up and run the VADER model
sia = SentimentIntensityAnalyzer()

df["vader_score"] = [sia.polarity_scores(text)["compound"] for text in df["Phrase"]]

# there are a total of 5 sentiment scores but polarity only gives us scores between -1 to 1
# need to map the scores from the dataset so that polarity can use it
def score_to_label(score):
    if score <= -0.6:
        return 0  # very negative
    elif score <= -0.2:
        return 1  # somewhat negative
    elif score < 0.2:
        return 2  # neutral
    elif score < 0.6:
        return 3  # somewhat positive
    else:
        return 4  # very positive

df["vader_pred"] = df["vader_score"].apply(score_to_label)

# View the predicted score
df.head()
````

### 🔧 Try It Yourself — Part 3

### In Your Response:

````python
# 1
print(classification_report(df["Sentiment"], df["vader_pred"]))
````

````python
# 2
cm = confusion_matrix(df["Sentiment"], df["vader_pred"])
print(cm)
````

### ✍️ Your Response: 🔧
1. The values for a sentiment score of 2 are much higher because many phrases in the dataset are short, so VADER classifies them as neutral. Since VADER analyzes individual words and simple rules, it often defaults to neutral when the sentiment is unclear.

## How TextBlob Works:
Unlike VADER, TextBlod utilizes tokenization to determine the sentiment of a phrase. Tokenization is great for breaking up large sentences and paragraphs. Because of this, TextBlob works better on longer sentences and is best suited for analyzing longer text documents such as articles or blogs.

Tokenization is the process of breaking text into smaller pieces called tokens, which are usually words, phrases, or sentences. For example, the sentence “Loved the movie!” might be split into tokens like ["loved", "the", "movie", "!"]. This step is important because computers need text to be structured in smaller units before they can analyze it—whether for tasks like counting words, understanding meaning, or performing sentiment analysis.

**Things to notice:**
- how does accuracy compare to the previous model?

````python

# set up and run the textblob model
df["textblob_score"] = [TextBlob(text).sentiment.polarity for text in df["Phrase"]]
df["textblob_pred"] = df["textblob_score"].apply(score_to_label)

# View the predicted score
df.head()
````
